```python
import requests
from bs4 import BeautifulSoup

# Function to scrape data from a given URL
def scrape_data(url):
    # Send a GET request to the URL
    response = requests.get(url)
    
    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content using Beautiful Soup
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # Extract text content from the webpage
        text = soup.get_text()
        
        # Return the scraped text
        return text
    else:
        print(f"Failed to fetch data from {url}")
        return None

# Function to fetch latest news for a company (e.g., Reliance Industries Ltd. - RIL)
def fetch_latest_news(company_name):
    # List to store dictionaries containing scraped data
    news_data = []
    
    # List of URLs to scrape data from (e.g., news websites, Twitter, etc.)
    urls = [
        "https://twitter.com/search?q=Reliance%20Industries%20Ltd&src=typed_query&f=live",  # Twitter
        "https://www.moneycontrol.com/news/tags/reliance-industries.html",  # Moneycontrol
        # Add more URLs as needed for other sources
    ]
    
    # Scrape data from each URL and add it to the news_data list
    for url in urls:
        # Scrape data from the URL
        scraped_text = scrape_data(url)
        
        # If data is scraped successfully, add it to the news_data list
        if scraped_text:
            news_data.append({
                "source": url,
                "text": scraped_text
            })
    
    # Return the list of dictionaries containing scraped data
    return news_data

# Fetch latest news for Reliance Industries Ltd. (RIL)
company_name = "Reliance Industries Ltd."
latest_news = fetch_latest_news(company_name)

# Print the list of dictionaries containing scraped data
for news_item in latest_news:
    print(news_item)
```
